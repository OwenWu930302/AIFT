```python
import yfinance as yf
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import statsmodels.api as sm
import inspect
import os
from datetime import datetime
import warnings
import seaborn as sns
from statsmodels.tsa.stattools import coint
from datetime import datetime, timedelta



# Suppress warnings
warnings.filterwarnings("ignore")

# 設定繪圖風格
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("pastel")
# Create a directory for saving results
if not os.path.exists("results"):
    os.makedirs("results")

# Constants
INITIAL_CAPITAL = 10000
# Using 5 diverse stocks from different sectors
TICKERS = ['AAPL', 'MSFT', 'JPM', 'PFE', 'XOM']
START_DATE = "2015-01-01"
END_DATE = "2023-01-01"

# 1. Fetch data for all five stocks
print("Fetching stock data...")
stock_data = {}
for ticker in TICKERS:
    stock_data[ticker] = yf.download(ticker, start=START_DATE, end=END_DATE)
    print(f"Downloaded data for {ticker}: {len(stock_data[ticker])} trading days")

# 2. LSTM Model for price prediction
class LSTMModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(input_size=1, hidden_size=32, num_layers=1, batch_first=True)
        self.fc = nn.Linear(32, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

def create_dataset(data, seq_len=30):
    """Creates sequence data for LSTM"""
    X, y = [], []
    for i in range(len(data)-seq_len):
        X.append(data[i:i+seq_len])
        y.append(data[i+seq_len])
    return np.array(X), np.array(y)

def train_lstm_model(data_scaled, seq_len=30, epochs=20):
    """Train LSTM model on scaled data"""
    X, y = create_dataset(data_scaled, seq_len)
    X = torch.tensor(X, dtype=torch.float32)
    y = torch.tensor(y, dtype=torch.float32)

    model = LSTMModel()
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

    for epoch in range(epochs):
        output = model(X)
        loss = criterion(output, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if epoch % 5 == 0:
            print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

    return model, X, y

# 3. Define investment strategies
def lump_sum_strategy(prices, initial_capital=INITIAL_CAPITAL):
    """Buy and hold strategy - invest all at the beginning"""
    if len(prices) == 0:
        return [initial_capital]

    shares = initial_capital / prices[0]
    portfolio_values = [shares * price for price in prices]

    # Ensure we have a proper list of values
    if not portfolio_values:
        return [initial_capital] * len(prices)

    return portfolio_values

def dca_strategy(prices, initial_capital=INITIAL_CAPITAL, intervals=12):
    """Dollar Cost Averaging - invest equal amounts at regular intervals"""
    if len(prices) == 0:
        return [initial_capital]

    period_investment = initial_capital / intervals
    shares = 0
    cash = initial_capital
    portfolio_values = []

    interval_days = max(1, len(prices) // intervals)

    for i, price in enumerate(prices):
        if i % interval_days == 0 and cash > 0:
            investment = min(period_investment, cash)
            shares += investment / price
            cash -= investment

        portfolio_value = shares * price + cash
        portfolio_values.append(portfolio_value)

    # Ensure we have a proper list of values
    if not portfolio_values:
        return [initial_capital] * len(prices)

    return portfolio_values

def lstm_strategy(model, scaled_data, scaler, initial_capital=INITIAL_CAPITAL, seq_len=30):
    """Trading strategy based on LSTM predictions"""
    try:
        price_data = scaler.inverse_transform(scaled_data)[:, 0]

        # Make predictions
        with torch.no_grad():
            X, _ = create_dataset(scaled_data, seq_len)
            X_tensor = torch.tensor(X, dtype=torch.float32)
            predicted = model(X_tensor).numpy()

        pred_price = scaler.inverse_transform(predicted)[:, 0]

        # Trading logic - buy when predicted price rise, sell when predicted price fall
        position = 0
        cash = initial_capital
        portfolio = [initial_capital]  # Start with initial capital

        for i in range(1, len(pred_price)):
            current_price = price_data[i+seq_len-1]  # Current actual price

            if pred_price[i] > pred_price[i-1] and position == 0:  # Predicted price rise
                position = cash / current_price
                cash = 0
            elif pred_price[i] < pred_price[i-1] and position > 0:  # Predicted price fall
                cash = position * current_price
                position = 0

            portfolio_value = cash + position * current_price
            portfolio.append(portfolio_value)

        # Ensure we have a proper list of values
        if not portfolio:
            return [initial_capital] * len(price_data)

        return portfolio
    except Exception as e:
        print(f"Error in LSTM strategy: {str(e)}")
        # Return a default portfolio of initial capital
        return [initial_capital] * len(scaled_data)

def moving_average_strategy(prices, short_window=20, long_window=50, initial_capital=INITIAL_CAPITAL):
    """Moving Average crossover strategy"""
    if len(prices) <= long_window:
        return [initial_capital] * len(prices)

    try:
        # Calculate moving averages
        df = pd.DataFrame(index=range(len(prices)))
        df['price'] = prices
        df['short_ma'] = df['price'].rolling(window=short_window).mean()
        df['long_ma'] = df['price'].rolling(window=long_window).mean()

        # Generate signals - using loc instead of chained assignment
        df['signal'] = 0.0
        df.loc[long_window:, 'signal'] = np.where(df['short_ma'][long_window:] > df['long_ma'][long_window:], 1.0, 0.0)
        df['position'] = df['signal'].diff()

        # Initialize portfolio metrics
        cash = initial_capital
        shares = 0
        portfolio_values = [initial_capital] * long_window  # Initialize with initial capital

        # Backtest the strategy
        for i in range(long_window, len(df)):
            price = df['price'][i]

            # Buy signal
            if df['position'][i] == 1.0:
                shares = cash / price
                cash = 0

            # Sell signal
            elif df['position'][i] == -1.0:
                cash = shares * price
                shares = 0

            portfolio_value = cash + shares * price
            portfolio_values.append(portfolio_value)

        # Ensure we have a proper list of values
        if not portfolio_values:
            return [initial_capital] * len(prices)

        return portfolio_values
    except Exception as e:
        print(f"Error in Moving Average strategy: {str(e)}")
        return [initial_capital] * len(prices)

# Enhanced moving average strategy with RSI for confirmation
def enhanced_ma_rsi_strategy(prices, short_window=20, long_window=50, rsi_period=14,
                            rsi_overbought=70, rsi_oversold=30, initial_capital=INITIAL_CAPITAL):
    """Enhanced Moving Average strategy with RSI confirmation"""
    if len(prices) <= max(long_window, rsi_period):
        return [initial_capital] * len(prices)

    try:
        # Calculate moving averages and RSI
        df = pd.DataFrame(index=range(len(prices)))
        df['price'] = prices
        df['short_ma'] = df['price'].rolling(window=short_window).mean()
        df['long_ma'] = df['price'].rolling(window=long_window).mean()

        # Calculate RSI
        delta = df['price'].diff()
        gain = delta.where(delta > 0, 0).fillna(0)
        loss = -delta.where(delta < 0, 0).fillna(0)

        avg_gain = gain.rolling(window=rsi_period).mean()
        avg_loss = loss.rolling(window=rsi_period).mean()

        # Handle division by zero
        avg_loss = avg_loss.replace(0, 0.00001)

        rs = avg_gain / avg_loss
        df['rsi'] = 100 - (100 / (1 + rs))

        # Generate signals
        df['ma_signal'] = 0.0
        df.loc[long_window:, 'ma_signal'] = np.where(df['short_ma'][long_window:] > df['long_ma'][long_window:], 1.0, 0.0)

        # Only take buy signals when RSI is not overbought, and sell signals when RSI is not oversold
        df['signal'] = 0.0

        # Use iloc instead of boolean indexing to avoid SettingWithCopyWarning
        for i in range(max(long_window, rsi_period), len(df)):
            if df.iloc[i]['ma_signal'] == 1.0 and df.iloc[i]['rsi'] < rsi_overbought:
                df.loc[i, 'signal'] = 1.0
            elif df.iloc[i]['ma_signal'] == 0.0 and df.iloc[i]['rsi'] > rsi_oversold:
                df.loc[i, 'signal'] = 0.0

        df['position'] = df['signal'].diff()

        # Initialize portfolio metrics
        cash = initial_capital
        shares = 0
        min_idx = max(long_window, rsi_period)
        portfolio_values = [initial_capital] * min_idx  # Initialize with initial capital

        # Backtest the strategy
        for i in range(min_idx, len(df)):
            price = df['price'][i]

            # Buy signal
            if df['position'][i] == 1.0:
                shares = cash / price
                cash = 0

            # Sell signal
            elif df['position'][i] == -1.0:
                cash = shares * price
                shares = 0

            portfolio_value = cash + shares * price
            portfolio_values.append(portfolio_value)

        # Ensure we have a proper list of values
        if not portfolio_values:
            return [initial_capital] * len(prices)

        return portfolio_values
    except Exception as e:
        print(f"Error in Enhanced MA-RSI strategy: {str(e)}")
        return [initial_capital] * len(prices)

# Helper function to filter parameters for a function
def filter_params_for_function(func, params):
    """
    Filter parameters to only include those accepted by the function
    """
    sig = inspect.signature(func)
    valid_params = {}
    for name, param in sig.parameters.items():
        if name in params:
            valid_params[name] = params[name]
    return valid_params

# 4. Temporal Validation Implementation - Fixed to filter function parameters
def temporal_validation(prices, strategy_func, train_years, test_year, trading_days_per_year=252, **strategy_params):
    """
    Implement temporal validation for strategy testing
    prices: full price history
    train_years: number of years to use for training
    test_year: number of years to use for testing
    """
    try:
        # Calculate cutoff points
        train_days = int(train_years * trading_days_per_year)
        test_days = int(test_year * trading_days_per_year)

        if len(prices) < train_days + test_days:
            return None  # Not enough data

        # Split data
        train_prices = prices[:train_days]
        test_prices = prices[train_days:train_days+test_days]

        if strategy_func == lstm_strategy:
            # Special handling for LSTM strategy which requires preprocessing
            scaler = MinMaxScaler()
            train_scaled = scaler.fit_transform(train_prices.reshape(-1, 1))

            # Train LSTM model on training data
            seq_len = strategy_params.get('seq_len', 30)
            model, _, _ = train_lstm_model(train_scaled, seq_len)

            # Prepare full dataset for prediction (need to include training for context)
            full_scaled = scaler.transform(prices[:train_days+test_days].reshape(-1, 1))

            # Get portfolio values from LSTM strategy
            portfolio_values = lstm_strategy(model, full_scaled, scaler, INITIAL_CAPITAL, seq_len)
            # Only keep the test period values
            test_portfolio_values = portfolio_values[-test_days:]
        else:
            # For standard strategies - Filter parameters to only include those accepted by the function
            filtered_params = filter_params_for_function(strategy_func, strategy_params)
            test_portfolio_values = strategy_func(test_prices, **filtered_params)

        # Calculate annual return
        if len(test_portfolio_values) > 0:
            annual_return = calculate_annual_return(test_portfolio_values[-1], INITIAL_CAPITAL, test_year)
        else:
            annual_return = 0

        return {
            'portfolio_values': test_portfolio_values,
            'annual_return': annual_return
        }
    except Exception as e:
        print(f"Error in temporal validation: {str(e)}")
        return {
            'portfolio_values': [INITIAL_CAPITAL] * int(test_year * trading_days_per_year),
            'annual_return': 0
        }

# 5. Pairs Trading Implementation
def calculate_hedge_ratio(price1, price2):
    """Calculate hedge ratio between two stock price series"""
    try:
        X = sm.add_constant(price1)
        model = sm.OLS(price2, X).fit()
        return model.params[1]
    except Exception as e:
        print(f"Error calculating hedge ratio: {str(e)}")
        # Return a default ratio of 1.0 as fallback
        return 1.0

def pairs_trading_strategy(stock1_prices, stock2_prices, z_threshold=2.0, initial_capital=INITIAL_CAPITAL):
    """Implement pairs trading strategy with improved error handling"""
    try:
        # Ensure same length
        min_len = min(len(stock1_prices), len(stock2_prices))
        if min_len == 0:
            print("Warning: Empty price data provided")
            return [initial_capital]

        stock1_prices = stock1_prices[:min_len]
        stock2_prices = stock2_prices[:min_len]

        # Calculate hedge ratio
        hedge_ratio = calculate_hedge_ratio(stock1_prices, stock2_prices)

        # Print debug info
        print(f"Hedge ratio: {hedge_ratio}")

        # Handle invalid hedge ratio
        if np.isnan(hedge_ratio) or np.isinf(hedge_ratio) or abs(hedge_ratio) > 100:
            print(f"Warning: Invalid hedge ratio: {hedge_ratio}, using 1.0 instead")
            hedge_ratio = 1.0

        # Calculate spread
        spread = stock2_prices - hedge_ratio * stock1_prices

        # Calculate z-score with improved error handling
        spread_mean = np.mean(spread)
        spread_std = np.std(spread)

        # Handle zero standard deviation
        if spread_std < 0.00001:
            print("Warning: Near-zero spread standard deviation, using default value")
            spread_std = 0.01

        z_score = (spread - spread_mean) / spread_std

        # Debug info
        print(f"Spread mean: {spread_mean}, std: {spread_std}")
        print(f"Z-score range: {np.min(z_score)} to {np.max(z_score)}")

        # Initialize portfolio
        cash = initial_capital
        stock1_shares = 0
        stock2_shares = 0
        portfolio_values = [initial_capital]

        # Backtest the strategy with additional safety checks
        for i in range(1, len(z_score)):
            # Skip invalid prices
            if np.isnan(stock1_prices[i]) or np.isnan(stock2_prices[i]):
                portfolio_values.append(portfolio_values[-1])
                continue

            # Trading logic
            if z_score[i-1] > z_threshold and z_score[i] <= z_threshold:
                # Close short position in stock1 and long position in stock2
                if stock1_shares < 0:
                    cash += stock1_shares * stock1_prices[i] # Buy back stock1 (close short)
                    cash += stock2_shares * stock2_prices[i] # Sell stock2 (close long)
                    stock1_shares = 0
                    stock2_shares = 0

            elif z_score[i-1] < -z_threshold and z_score[i] >= -z_threshold:
                # Close long position in stock1 and short position in stock2
                if stock1_shares > 0:
                    cash += stock1_shares * stock1_prices[i] # Sell stock1 (close long)
                    cash += stock2_shares * stock2_prices[i] # Buy back stock2 (close short)
                    stock1_shares = 0
                    stock2_shares = 0

            elif z_score[i] > z_threshold and stock1_shares == 0:
                # Open short position in stock1 and long position in stock2
                trade_amount = cash / 2
                stock1_shares = -trade_amount / stock1_prices[i] # Short stock1
                stock2_shares = trade_amount / stock2_prices[i] # Long stock2
                cash += trade_amount # Cash from shorting stock1
                cash -= trade_amount # Cash used to buy stock2

            elif z_score[i] < -z_threshold and stock1_shares == 0:
                # Open long position in stock1 and short position in stock2
                trade_amount = cash / 2
                stock1_shares = trade_amount / stock1_prices[i] # Long stock1
                stock2_shares = -trade_amount / stock2_prices[i] # Short stock2
                cash -= trade_amount # Cash used to buy stock1
                cash += trade_amount # Cash from shorting stock2

            # Calculate current portfolio value
            current_value = cash + stock1_shares * stock1_prices[i] + stock2_shares * stock2_prices[i]
            portfolio_values.append(current_value)

        # Final check for data quality
        if len(portfolio_values) < min_len:
            print(f"Warning: Portfolio values length ({len(portfolio_values)}) less than price data ({min_len})")
            # Extend with last value if needed
            portfolio_values.extend([portfolio_values[-1]] * (min_len - len(portfolio_values)))

        # Print summary statistics for debugging
        if len(portfolio_values) > 0:
            print(f"Final portfolio value: {portfolio_values[-1]}")
            print(f"Return: {((portfolio_values[-1] / initial_capital) - 1) * 100:.2f}%")

        return portfolio_values
    except Exception as e:
        print(f"Error in pairs trading strategy: {str(e)}")
        import traceback
        traceback.print_exc()  # Print full stack trace
        return [initial_capital] * min(len(stock1_prices), len(stock2_prices))


# Function to implement temporal validation for pairs trading
def pairs_trading_temporal_validation(stock1_prices, stock2_prices, train_years, trading_days_per_year=252,
                                      z_threshold=2.0, initial_capital=INITIAL_CAPITAL):
    """
    Implement temporal validation for pairs trading

    Parameters:
    - stock1_prices: Price data for first stock
    - stock2_prices: Price data for second stock
    - train_years: Number of years to use for training
    - trading_days_per_year: Number of trading days in a year
    - z_threshold: Z-score threshold for trading signals
    - initial_capital: Initial investment amount

    Returns:
    - Dictionary with test results
    """
    try:
        # Calculate training period in days
        train_days = int(train_years * trading_days_per_year)

        # Ensure there's enough data
        min_len = min(len(stock1_prices), len(stock2_prices))
        if min_len <= train_days:
            return None  # Not enough data

        # Split data into training and testing
        stock1_train = stock1_prices[:train_days]
        stock1_test = stock1_prices[train_days:]

        stock2_train = stock2_prices[:train_days]
        stock2_test = stock2_prices[train_days:]

        # Calculate hedge ratio on training data
        X = sm.add_constant(stock1_train)
        model = sm.OLS(stock2_train, X).fit()
        hedge_ratio = model.params[1]

        # Calculate spread on training data
        train_spread = stock2_train - hedge_ratio * stock1_train
        spread_mean = np.mean(train_spread)
        spread_std = np.std(train_spread)

        # Handle zero standard deviation
        if spread_std == 0:
            spread_std = 0.00001

        # Calculate the spread on test data using parameters from training
        test_spread = stock2_test - hedge_ratio * stock1_test
        z_score = (test_spread - spread_mean) / spread_std

        # Initialize portfolio
        cash = initial_capital
        stock1_shares = 0
        stock2_shares = 0
        portfolio_values = [initial_capital]

        # Backtest on test data
        for i in range(1, len(z_score)):
            # Trading logic
            if z_score[i-1] > z_threshold and z_score[i] <= z_threshold:
                # Close short position
                if stock1_shares < 0:
                    cash += stock1_shares * stock1_test[i]
                    cash += stock2_shares * stock2_test[i]
                    stock1_shares = 0
                    stock2_shares = 0

            elif z_score[i-1] < -z_threshold and z_score[i] >= -z_threshold:
                # Close long position
                if stock1_shares > 0:
                    cash += stock1_shares * stock1_test[i]
                    cash += stock2_shares * stock2_test[i]
                    stock1_shares = 0
                    stock2_shares = 0

            elif z_score[i] > z_threshold and stock1_shares == 0:
                # Open short position in stock1, long in stock2
                trade_amount = cash / 2
                stock1_shares = -trade_amount / stock1_test[i]
                stock2_shares = trade_amount / stock2_test[i]
                cash += trade_amount
                cash -= trade_amount

            elif z_score[i] < -z_threshold and stock1_shares == 0:
                # Open long position in stock1, short in stock2
                trade_amount = cash / 2
                stock1_shares = trade_amount / stock1_test[i]
                stock2_shares = -trade_amount / stock2_test[i]
                cash -= trade_amount
                cash += trade_amount

            # Calculate portfolio value
            portfolio_value = cash + stock1_shares * stock1_test[i] + stock2_shares * stock2_test[i]
            portfolio_values.append(portfolio_value)

        # Calculate test period in years
        test_years = len(stock1_test) / trading_days_per_year

        # Calculate annual return
        annual_return = calculate_annual_return(portfolio_values[-1], initial_capital, test_years)

        return {
            'portfolio_values': portfolio_values,
            'annual_return': annual_return,
            'test_years': test_years
        }
    except Exception as e:
        print(f"Error in pairs trading temporal validation: {str(e)}")
        test_days = len(stock1_prices) - train_days if len(stock1_prices) > train_days else 1
        test_years = test_days / trading_days_per_year
        return {
            'portfolio_values': [initial_capital] * test_days,
            'annual_return': 0,
            'test_years': test_years
        }

# Helper function to calculate annualized returns - FIXED to handle array inputs properly
def calculate_annual_return(final_value, initial_value, years):
    """Calculate annualized return, handling both scalar and array inputs"""
    try:
        # If final_value is an array, take the first (or last) element
        if isinstance(final_value, (list, np.ndarray)):
            if len(final_value) > 0:
                final_value = float(final_value[-1])  # Take the last value if it's a list/array
            else:
                return 0  # Empty array, return 0

        # If initial_value is an array, take the first (or first) element
        if isinstance(initial_value, (list, np.ndarray)):
            if len(initial_value) > 0:
                initial_value = float(initial_value[0])  # Take the first value if it's a list/array
            else:
                return 0  # Empty array, return 0

        # Convert to float to ensure scalar calculation
        final_value = float(final_value)
        initial_value = float(initial_value)
        years = float(years)

        # Handle edge cases
        if years <= 0 or initial_value <= 0 or final_value <= 0:
            return 0

        # Calculate annual return
        return (final_value / initial_value) ** (1 / years) - 1
    except Exception as e:
        print(f"Error calculating annual return: {str(e)}")
        return 0

# FIXED: Function to calculate risk metrics with improved error handling
def calculate_hedge_ratio(price1, price2):
    """Calculate hedge ratio between two stock price series using OLS regression"""
    try:
        X = sm.add_constant(price1)
        model = sm.OLS(price2, X).fit()
        return model.params[1]
    except Exception as e:
        print(f"Error calculating hedge ratio: {str(e)}")
        return 1.0

def pairs_trading_strategy(stock1_prices, stock2_prices, z_threshold=2.0, initial_capital=10000):
    # 1. 對齊長度
    min_len = min(len(stock1_prices), len(stock2_prices))
    stock1_prices = np.array(stock1_prices[:min_len])
    stock2_prices = np.array(stock2_prices[:min_len])

    # 2. 計算對沖比率（hedge ratio）
    X = sm.add_constant(stock1_prices)
    model = sm.OLS(stock2_prices, X).fit()
    hedge_ratio = model.params[1]

    # 3. 計算 spread 及 z-score
    spread = stock2_prices - hedge_ratio * stock1_prices
    spread_mean = np.mean(spread)
    spread_std = np.std(spread)
    if spread_std == 0:
        spread_std = 1e-6
    zscore = (spread - spread_mean) / spread_std

    # 4. 回測
    cash = initial_capital
    stock1_shares = 0
    stock2_shares = 0
    portfolio_values = []

    for i in range(min_len):
        # 平倉條件
        if stock1_shares != 0 and abs(zscore[i]) < 0.5:
            cash += stock1_shares * stock1_prices[i]
            cash += stock2_shares * stock2_prices[i]
            stock1_shares = 0
            stock2_shares = 0

        # 開倉條件
        elif stock1_shares == 0:
            if zscore[i] > z_threshold:
                # 做空stock1，做多stock2
                half_cash = cash / 2
                stock1_shares = -half_cash / stock1_prices[i]
                stock2_shares = half_cash / stock2_prices[i]
                cash = 0
            elif zscore[i] < -z_threshold:
                # 做多stock1，做空stock2
                half_cash = cash / 2
                stock1_shares = half_cash / stock1_prices[i]
                stock2_shares = -half_cash / stock2_prices[i]
                cash = 0

        # 計算當前投資組合價值
        portfolio_value = cash + stock1_shares * stock1_prices[i] + stock2_shares * stock2_prices[i]
        portfolio_values.append(portfolio_value)

    return portfolio_values

def ensure_1d_array(arr):
    """確保陣列是1D並改進錯誤處理"""
    try:
        # 檢查輸入是否為None或空
        if arr is None or (hasattr(arr, '__len__') and len(arr) == 0):
            print("Warning: Input array is empty or None")
            return np.array([INITIAL_CAPITAL, INITIAL_CAPITAL * 1.01])  # 返回有變化的陣列

        # 轉換為numpy陣列
        if isinstance(arr, (list, np.ndarray, pd.Series)):
            arr = np.asarray(arr)

            # 處理多維陣列
            if arr.ndim > 1:
                print(f"Converting {arr.ndim}D array to 1D")
                arr = arr.flatten()

            # 檢查NaN值
            if np.isnan(arr).any():
                print(f"Warning: Array contains NaN values, replacing with 0")
                arr = np.nan_to_num(arr)

            # 檢查陣列值是否變化
            if len(arr) > 1:
                if np.max(arr) == np.min(arr):
                    print("WARNING: All array values are identical! Adding variation for visualization.")
                    # 添加小變化以便可視化
                    arr = arr + np.linspace(0, arr[0]*0.01, len(arr))
                else:
                    print(f"Array range: min={np.min(arr):.2f}, max={np.max(arr):.2f}, length={len(arr)}")

            return arr
        else:
            print(f"Warning: Input is not an array-like object: {type(arr)}")
            return np.array([INITIAL_CAPITAL, INITIAL_CAPITAL * 1.01])
    except Exception as e:
        print(f"Error in ensure_1d_array: {str(e)}")
        return np.array([INITIAL_CAPITAL, INITIAL_CAPITAL * 1.01])


def default_risk_metrics():
    """Return default risk metrics when calculation fails"""
    return {
        'annualized_return': 0,
        'annualized_volatility': 0,
        'sharpe_ratio': 0,
        'max_drawdown': 0,
        'sortino_ratio': 0
    }

# Main execution
def main():
    results = {}
    pairs_results = {}

    # Problem 1: Evaluate strategies for each stock
    for ticker in TICKERS:
        print(f"\n=== Analyzing {ticker} ===")

        # Get price data
        prices = stock_data[ticker]['Close'].values
        trading_years = (len(prices) - 1) / 252  # Approximate years of data

        # Prepare data for LSTM
        scaler = MinMaxScaler()
        data_scaled = scaler.fit_transform(prices.reshape(-1, 1))

        # Train LSTM model
        print("Training LSTM model...")
        seq_len = 30
        model, X, y = train_lstm_model(data_scaled, seq_len, epochs=20)

        # Implement strategies
        print("Implementing investment strategies...")
        lump_sum_values = lump_sum_strategy(prices)
        dca_values = dca_strategy(prices)
        lstm_values = lstm_strategy(model, data_scaled, scaler, INITIAL_CAPITAL, seq_len)
        ma_values = moving_average_strategy(prices)
        enhanced_ma_values = enhanced_ma_rsi_strategy(prices)

        # Calculate annual returns
        lump_sum_annual = calculate_annual_return(lump_sum_values[-1], INITIAL_CAPITAL, trading_years)
        dca_annual = calculate_annual_return(dca_values[-1], INITIAL_CAPITAL, trading_years)
        lstm_annual = calculate_annual_return(lstm_values[-1], INITIAL_CAPITAL, trading_years)
        ma_annual = calculate_annual_return(ma_values[-1], INITIAL_CAPITAL, trading_years)
        enhanced_ma_annual = calculate_annual_return(enhanced_ma_values[-1], INITIAL_CAPITAL, trading_years)

        # Calculate risk metrics
        lump_sum_risk = calculate_risk_metrics(lump_sum_values)
        dca_risk = calculate_risk_metrics(dca_values)
        lstm_risk = calculate_risk_metrics(lstm_values)
        ma_risk = calculate_risk_metrics(ma_values)
        enhanced_ma_risk = calculate_risk_metrics(enhanced_ma_values)

        print(f"Lump Sum Annual Return: {lump_sum_annual:.2%}")
        print(f"DCA Annual Return: {dca_annual:.2%}")
        print(f"LSTM Strategy Annual Return: {lstm_annual:.2%}")
        print(f"Moving Average Annual Return: {ma_annual:.2%}")
        print(f"Enhanced MA-RSI Annual Return: {enhanced_ma_annual:.2%}")

        # Store results
        results[ticker] = {
            'lump_sum': {'values': lump_sum_values, 'annual_return': lump_sum_annual, 'risk': lump_sum_risk},
            'dca': {'values': dca_values, 'annual_return': dca_annual, 'risk': dca_risk},
            'lstm': {'values': lstm_values, 'annual_return': lstm_annual, 'risk': lstm_risk},
            'moving_average': {'values': ma_values, 'annual_return': ma_annual, 'risk': ma_risk},
            'enhanced_ma': {'values': enhanced_ma_values, 'annual_return': enhanced_ma_annual, 'risk': enhanced_ma_risk},
            'temporal_validation': {}
        }

        # Temporal validation
        print("Performing temporal validation...")
        # For each strategy
        for strategy_name, strategy_func in [
            ('moving_average', moving_average_strategy),
            ('enhanced_ma', enhanced_ma_rsi_strategy),
            ('lstm', lstm_strategy)  # Including LSTM for comparison
        ]:
            results[ticker]['temporal_validation'][strategy_name] = {}

            # Different training/testing splits
            for train_years in range(1, 5):  # 1 to 4 years of training
                # Only pass seq_len to lstm_strategy
                strategy_params = {}
                if strategy_func == lstm_strategy:
                    strategy_params['seq_len'] = seq_len

                tv_result = temporal_validation(
                    prices,
                    strategy_func,
                    train_years,
                    1,  # Test on 1 year
                    **strategy_params
                )

                if tv_result:
                    results[ticker]['temporal_validation'][strategy_name][train_years] = tv_result
                    print(f"  TV{train_years} {strategy_name} annual return: {tv_result['annual_return']:.2%}")

        # Visualize strategies - using ensure_1d_array to prevent plotting errors
        try:
            plt.figure(figsize=(12, 6))

            # Ensure arrays are 1D for plotting
            lump_sum_values_1d = ensure_1d_array(lump_sum_values)
            dca_values_1d = ensure_1d_array(dca_values)
            lstm_values_1d = ensure_1d_array(lstm_values)
            ma_values_1d = ensure_1d_array(ma_values)
            enhanced_ma_values_1d = ensure_1d_array(enhanced_ma_values)

            plt.plot(lump_sum_values_1d, label=f'Lump Sum ({lump_sum_annual:.2%})')
            plt.plot(dca_values_1d, label=f'DCA ({dca_annual:.2%})')
            plt.plot(lstm_values_1d, label=f'LSTM Strategy ({lstm_annual:.2%})')
            plt.plot(ma_values_1d, label=f'Moving Average ({ma_annual:.2%})')
            plt.plot(enhanced_ma_values_1d, label=f'Enhanced MA-RSI ({enhanced_ma_annual:.2%})')
            plt.title(f"Investment Strategies for {ticker}")
            plt.xlabel("Trading Days")
            plt.ylabel("Portfolio Value ($)")
            plt.legend()
            plt.grid(True)
            plt.savefig(f"results/{ticker}_strategies.png")
            plt.show()
        except Exception as e:
            print(f"Error plotting strategy results: {str(e)}")

    # Problem 2: Pairs Trading
    print("\n=== Evaluating Pairs Trading ===")

    # Try all possible pairs
    for i in range(len(TICKERS)):
        for j in range(i+1, len(TICKERS)):
            ticker1 = TICKERS[i]
            ticker2 = TICKERS[j]
            pair_name = f"{ticker1}-{ticker2}"

            print(f"\nAnalyzing pair: {pair_name}")

            # Get prices
            stock1_prices = stock_data[ticker1]['Close'].values
            stock2_prices = stock_data[ticker2]['Close'].values

            # Calculate pairs trading
            try:
                portfolio_values = pairs_trading_strategy(stock1_prices, stock2_prices)

                # Calculate annual return
                min_len = min(len(stock1_prices), len(stock2_prices))
                trading_years = min_len / 252
                annual_return = calculate_annual_return(portfolio_values[-1], INITIAL_CAPITAL, trading_years)

                # Calculate risk metrics
                risk_metrics = calculate_risk_metrics(portfolio_values)

                print(f"Pairs Trading Annual Return: {annual_return:.2%}")
                print(f"Sharpe Ratio: {risk_metrics['sharpe_ratio']:.2f}")
                print(f"Maximum Drawdown: {risk_metrics['max_drawdown']:.2%}")

                # Store results
                pairs_results[pair_name] = {
                    'portfolio_values': portfolio_values,
                    'annual_return': annual_return,
                    'risk_metrics': risk_metrics,
                    'temporal_validation': {}
                }

                # Temporal validation for pairs trading
                print("Performing temporal validation for pairs trading...")

                for train_years in range(1, 5):  # 1 to 4 years of training
                    # Use the pairs trading temporal validation function
                    tv_result = pairs_trading_temporal_validation(
                        stock1_prices,
                        stock2_prices,
                        train_years
                    )

                    if tv_result:
                        pairs_results[pair_name]['temporal_validation'][train_years] = tv_result
                        print(f"  TV{train_years} annual return: {tv_result['annual_return']:.2%}")

                # Visualize pairs trading - ensure 1D array for plotting
                try:
                    plt.figure(figsize=(12, 6))
                    portfolio_values_1d = ensure_1d_array(portfolio_values)
                    plt.plot(portfolio_values_1d, label=f'Pairs Trading ({annual_return:.2%})')
                    plt.title(f"Pairs Trading: {pair_name}")
                    plt.xlabel("Trading Days")
                    plt.ylabel("Portfolio Value ($)")
                    plt.legend()
                    plt.grid(True)
                    plt.savefig(f"results/{pair_name}_pairs_trading.png")
                    plt.show()

                    # Plot the spread and z-score
                    plt.figure(figsize=(12, 6))

                    # Calculate hedge ratio, spread, and z-score for visualization
                    hedge_ratio = calculate_hedge_ratio(stock1_prices, stock2_prices)
                    spread = stock2_prices - hedge_ratio * stock1_prices

                    # Handle zero standard deviation
                    spread_std = np.std(spread)
                    if spread_std == 0:
                        spread_std = 0.00001

                    z_score = (spread - np.mean(spread)) / spread_std

                    plt.subplot(2, 1, 1)
                    plt.plot(spread)
                    plt.title(f"{pair_name} Spread")
                    plt.axhline(y=np.mean(spread), color='r', linestyle='-')
                    plt.axhline(y=np.mean(spread) + 2*np.std(spread), color='g', linestyle='--')
                    plt.axhline(y=np.mean(spread) - 2*np.std(spread), color='g', linestyle='--')

                    plt.subplot(2, 1, 2)
                    plt.plot(z_score)
                    plt.title(f"{pair_name} Z-Score")
                    plt.axhline(y=0, color='r', linestyle='-')
                    plt.axhline(y=2, color='g', linestyle='--')
                    plt.axhline(y=-2, color='g', linestyle='--')

                    plt.tight_layout()
                    plt.savefig(f"results/{pair_name}_spread_zscore.png")
                    plt.show()
                except Exception as e:
                    print(f"Error plotting pairs trading results: {str(e)}")

            except Exception as e:
                print(f"Error analyzing pair {pair_name}: {str(e)}")

    # Print summary of results
    print("\n=== SUMMARY OF RESULTS ===")

    # Create summary table for individual stock strategies
    print("\nIndividual Stock Strategies:")
    summary_data = []
    for ticker in TICKERS:
        ticker_data = [
            ticker,
            f"{results[ticker]['lump_sum']['annual_return']:.2%}",
            f"{results[ticker]['dca']['annual_return']:.2%}",
            f"{results[ticker]['moving_average']['annual_return']:.2%}",
            f"{results[ticker]['enhanced_ma']['annual_return']:.2%}",
            f"{results[ticker]['lstm']['annual_return']:.2%}"
        ]
        summary_data.append(ticker_data)

    summary_df = pd.DataFrame(
        summary_data,
        columns=['Ticker', 'Lump Sum', 'DCA', 'Moving Average', 'Enhanced MA-RSI', 'LSTM']
    )
    print(summary_df.to_string(index=False))

    # Save summary table to CSV
    summary_df.to_csv("results/stock_strategies_summary.csv", index=False)

    # Create summary table for pairs trading
    print("\nPairs Trading Strategies:")
    pairs_data = []
    for pair_name, pair_result in pairs_results.items():
        pair_data = [
            pair_name,
            f"{pair_result['annual_return']:.2%}",
            f"{pair_result['risk_metrics']['sharpe_ratio']:.2f}",
            f"{pair_result['risk_metrics']['max_drawdown']:.2%}"
        ]
        pairs_data.append(pair_data)

    pairs_df = pd.DataFrame(
        pairs_data,
        columns=['Pair', 'Annual Return', 'Sharpe Ratio', 'Max Drawdown']
    )
    print(pairs_df.to_string(index=False))

    # Save pairs summary to CSV
    pairs_df.to_csv("results/pairs_trading_summary.csv", index=False)

    # Create temporal validation summary
    for train_years in range(1, 5):
        print(f"\nTemporal Validation (TV{train_years}) Summary:")
        tv_data = []

        for ticker in TICKERS:
            ticker_tv = []
            for strategy in ['moving_average', 'enhanced_ma', 'lstm']:
                if (strategy in results[ticker]['temporal_validation'] and
                    train_years in results[ticker]['temporal_validation'][strategy] and
                    results[ticker]['temporal_validation'][strategy][train_years] is not None):
                    annual_return = results[ticker]['temporal_validation'][strategy][train_years]['annual_return']
                    ticker_tv.append(f"{annual_return:.2%}")
                else:
                    ticker_tv.append("N/A")

            tv_data.append([ticker] + ticker_tv)

        tv_df = pd.DataFrame(
            tv_data,
            columns=['Ticker', 'Moving Average', 'Enhanced MA-RSI', 'LSTM']
        )
        print(tv_df.to_string(index=False))

        # Save TV summary to CSV
        tv_df.to_csv(f"results/tv{train_years}_summary.csv", index=False)

    # Pairs trading temporal validation summary
    for train_years in range(1, 5):
        print(f"\nPairs Trading Temporal Validation (TV{train_years}) Summary:")
        tv_pairs_data = []

        for pair_name, pair_result in pairs_results.items():
            if (train_years in pair_result['temporal_validation'] and
                pair_result['temporal_validation'][train_years] is not None):
                annual_return = pair_result['temporal_validation'][train_years]['annual_return']
                tv_pairs_data.append([pair_name, f"{annual_return:.2%}"])
            else:
                tv_pairs_data.append([pair_name, "N/A"])

        tv_pairs_df = pd.DataFrame(
            tv_pairs_data,
            columns=['Pair', 'Annual Return']
        )
        print(tv_pairs_df.to_string(index=False))

        # Save pairs TV summary to CSV
        tv_pairs_df.to_csv(f"results/pairs_tv{train_years}_summary.csv", index=False)

if __name__ == "__main__":
    main()
```
